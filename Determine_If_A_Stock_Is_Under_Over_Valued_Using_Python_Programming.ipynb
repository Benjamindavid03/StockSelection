```python
# Finding undervalued stocks using this simple technique
# Using Profit to Earning P/E Ratio
#Concept : if company earnings = 4$ per share and the company is trading at 100$, P/E = 100/4 = 25, The stock is trading at 25 times the earnings
#get the P/E ratio of the particular sector
# get the P/E ratio of the stock's individually
```

```python
df = pd.read_csv('MW-NIFTY-MIDCAP-50-20-Aug-2021.csv');
df = pd.read_csv('MW-NIFTY-NEXT-50-20-Aug-2021.csv');
df = pd.read_csv('MW-NIFTY-50-20-Aug-2021.csv');
df
```

```python
#set the index to the Stock Ticker like ONGC
df = df.set_index('SYMBOL \n')
```

```python
#calculate the mean P/E ratio for the IT stocks in the dataset
PE_Ratio_Mean = df.PERatio.mean()
PE_Ratio_Mean
```

```python
#calculate and print the fare market value
#create a new column in the dataset for the fare_market_value
df['fare_market_value'] = PE_Ratio_Mean * df['Earnings_Per_Share'];
df
```

```python
# If the current_price of the stock is lesser than the newly created fare_market_value, then the particular stock is undervalued
df['Ratio'] =  df['current_price'] / df['fare_market_value'];
```

```python
#create a new column to store a label for over valued or under labeled
df['valued'] = np.where(df['Ratio'] <1.0, 'Under Valued', 'Fair or Over Valued');
df
```

```python
#create a column to show the percentage that stock is over valued or undervalued
df['value_percent'] = abs(df['Ratio'] - 1) * 100
```

```python
#show only a list of stocks that are under valued 
#prints the stock tickers that only under valued
df[df.valued == 'Under Valued'].index
```
